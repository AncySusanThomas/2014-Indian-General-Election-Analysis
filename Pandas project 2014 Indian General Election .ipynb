```python
election_data=data.fillna(0)
```

# Analyzation Between PC And State

```python
pcs=election_data['PC'].groupby(election_data['state/ut']).count()
```

```python
pcs
```

# Diagramatic Representation Of Above Analysis

```python
mylabels=['Andaman & Nicobar Islands','Andhra Pradesh','Arunachal Pradesh','Assam','Bihar','Chandigarh','Dadra & Nagar Haveli','Daman & Diu','Goa', 'Gujarat','Haryana','Himachal Pradesh','Jammu & Kashmir','Jharkhand','Karnataka','Kerala','Lakshadweep','Madhya Pradesh','Maharashtra','Manipur','Meghalaya','Mizoram','NCT OF Delhi','Nagaland','Puducherry','Punjab','Rajasthan','Sikkim','Tamil Nadu','Tripura','Uttar Pradesh','Uttarakhand','West Bengal']
plt.figure(figsize=(20,40))
plt.pie(pcs,labels=mylabels,autopct='%1.1f%%')
plt.legend()
plt.show()
```

**From the above analysis and visualization,UttarPradesh has the high number of PC.**

# Analyzation Between State and Votes

```python
state_vote=election_data['total_votes'].groupby(election_data['state/ut']).sum()
```

```python
state_vote
```

#  Diagramatic Representation Of Above Analysis

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
plt.grid()
sns.lineplot(state_vote)
```

**From above analysis we can understand that most number of votes were submited in UttarPradesh**

#  Analyzation Between Party And Votes

```python
party_vote=election_data['total_votes'].groupby(election_data['party']).sum()
```

```python
party_vote
```

#  Diagramatic Representation Of Above Analysis

```python
plt.figure(figsize=(70,25))
plt.xticks(rotation=90)
plt.grid()
sns.lineplot(party_vote)
```

**From above analysis All India Trinamool Congress got most number of votes**

#  Analyzation Between Votes And Candidates


```python
cand_vote=election_data['candidate_name'].groupby(election_data['total_votes']).sum()
```

```python
s=cand_vote.head(50)
s
```

#  Diagramatic Representation Of Above Analysis

```python
plt.figure(figsize=(20,25))
plt.xticks(rotation=90)
plt.grid()
sns.lineplot(s)
```

**From above analysis Jose K.Mani got most number of votes**
